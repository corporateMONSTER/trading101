```python
df_final=df.iloc[:,7]
```

```python
df_final=pd.DataFrame(df_final)
```

```python
df_final.columns
```

```python
rt=[]
for i in range(len(df)-1):
    num=df_final.iloc[i+1].values
    den=df_final.iloc[i].values
    print(num,den)
    val=np.log(num[0]/den[0])
    rt.append(val)
```

```python
for i in range(len(rt)):
    rt[i]=rt[i]*100
    
```

```python
len(rt)
```

```python
tot=0
length=len(rt)
for i in rt:
    tot=tot+i
print(tot/length)
```

```python
statistics.stdev(rt)
```
