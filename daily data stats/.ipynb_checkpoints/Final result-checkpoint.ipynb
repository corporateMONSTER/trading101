```python
cols=df.columns
cols=cols[1:-1]
print(cols)
```

```python
df_dict={}
for col in cols:
    rt=[]
    for i in range(len(df[col])-1):
        num=df[col].iloc[i+1]
        den=df[col].iloc[i]
        val=np.log(num/den)
        val_final=val*100
        rt.append(val_final)

    average=statistics.mean(rt)
    std=statistics.stdev(rt)
#     print("Average is ",col ," is ",average)
#     print("Standard deviation is ",col ," is ",std)
#     print("upper 1 avg+std is ",average+std)
#     print("lower 1 avg-std is ",average-std)
#     print("upper 2 avg+2std is ",average+(2*std))
#     print("lower 2 avg-2std is ",average-(2*std))
    df_dict[col]=[average+std,average-std,average+(2*std),average-(2*std)]
    



```
