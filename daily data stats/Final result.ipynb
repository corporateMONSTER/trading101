```python
import pandas as pd
import numpy as np
import statistics
```

```python
df=pd.read_excel('C:/Users/jayasuriya/Desktop/mission stocks/all data/Final data.xlsx')
```

```python
cols=df.columns
cols=cols[1:-1]
print(cols)
```

```python
df_dict={}
for col in cols:
    rt=[]
    for i in range(len(df[col])-1):
        num=df[col].iloc[i+1]
        den=df[col].iloc[i]
        val=np.log(num/den)
        val_final=val*100
        rt.append(val_final)

    average=statistics.mean(rt)
    std=statistics.stdev(rt)
    df_dict[col]=[(df[col][0]*np.exp((average+std)/100)),(df[col][0]*np.exp((average-std)/100)),(df[col][0]*np.exp((average+(2*std))/100)),(df[col][0]*np.exp((average-(2*std))/100)),(df[col][0]*np.exp((average+(3*std))/100)),(df[col][0]*np.exp((average-(3*std))/100))]
    



```
